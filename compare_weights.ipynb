```python
import torch
import matplotlib.pyplot as plt
from model_v2_compatible import SeqNN
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model_84 = SeqNN()
# the entire model
model_84.load_state_dict(torch.load("/scratch1/smaruj/train_pytorch_akita/mouse_models/v2_train_from_scratch_monitor/model_epoch_84.pth", map_location=device))
model_84.to(device)
model_84.eval()
```

```python
model_1 = SeqNN()
# the entire model
model_1.load_state_dict(torch.load("/scratch1/smaruj/train_pytorch_akita/mouse_models/v2_train_from_scratch_monitor/model_epoch_1.pth", map_location=device))
model_1.to(device)
model_1.eval()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

##  1st Conv

```python
weights_model_1 = model_1.conv_block_1.conv.weight.detach().cpu().numpy()  # [128, 4, 15]
weights_model_84 = model_84.conv_block_1.conv.weight.detach().cpu().numpy()
```

```python
# Plot overlapping histograms
plt.figure(figsize=(10, 6))
plt.hist(weights_model_1.flatten(), bins=50, alpha=0.5, label="Epoch 1", color='blue')
plt.hist(weights_model_84.flatten(), bins=50, alpha=0.5, label="Epoch 84", color='orange')

# Add titles and labels
plt.title("Weight Distribution Comparison")
plt.xlabel("Weight Value")
plt.ylabel("Frequency")
plt.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

```python
# 2. Plot the Difference in Weights
weight_diff = np.abs(weights_model_84 - weights_model_1)
```

```python

# Plot the absolute differences
plt.figure(figsize=(10, 6))
plt.hist(weight_diff.flatten(), bins=50, alpha=0.7, color='orange', label="Weight Changes")
plt.title("Absolute Weight Changes Between Epoch 1 and Epoch 84")
plt.xlabel("Weight Change")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
vmin = -4
vmax = 2
```

```python
# 3. Plot Individual Filters for Model 1 and Model 84 (Visualize First Few Filters)
# Select a few filters to visualize (e.g., the first 5 filters)
n_filters = 5
fig, axes = plt.subplots(n_filters, 2, figsize=(10, 10))

for i in range(n_filters):
    axes[i, 0].imshow(weights_model_1[i, :, :], cmap='viridis', vmin=vmin, vmax=vmax)  # Visualizing the first channel of filter i
    axes[i, 0].set_title(f"Model 1 - Filter {i + 1}")
    axes[i, 1].imshow(weights_model_84[i, :, :], cmap='viridis', vmin=vmin, vmax=vmax)  # Visualizing the first channel of filter i
    axes[i, 1].set_title(f"Model 84 - Filter {i + 1}")

plt.tight_layout()
plt.show()
```
