```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV
csv_path = "/scratch1/smaruj/Akita_pytorch_models/finetuned/mouse_models/Monahan2019_ORC/losses/Akita_v2_mouse_Monahan2019_ORC_model0_finetuned.csv"
df = pd.read_csv(csv_path)

# Split by stage
head_df = df[df['Stage'] == 'head']
full_df = df[df['Stage'] == 'full']
```

```python
plt.figure(figsize=(8,5))

# Plot head stage
plt.plot(head_df['Epoch'], head_df['Train Loss'], label='Head - Train', marker='o')
plt.plot(head_df['Epoch'], head_df['Validation Loss'], label='Head - Val', marker='o')

# Plot full model stage
plt.plot(full_df['Epoch'] + head_df['Epoch'].max(), full_df['Train Loss'], label='Full - Train', marker='x')
plt.plot(full_df['Epoch'] + head_df['Epoch'].max(), full_df['Validation Loss'], label='Full - Val', marker='x')

plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('Fine-tuning Loss: Monahan2019_ORC (Model 0)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
