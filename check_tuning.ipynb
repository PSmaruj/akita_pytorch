```python
import optuna
import matplotlib.pyplot as plt
```

```python
study = optuna.load_study(
    study_name="shuffled_ratio_tuning",
    storage="sqlite:///optuna_shuffled_ratio.db"
)
```

```python
df = study.trials_dataframe()
```

```python
plt.scatter(df['params_shuffled_ratio'], df['value'])
plt.xlabel("Shuffled Ratio")
plt.ylabel("Validation MSE Loss")
plt.title("Optuna: Shuffled Ratio vs. Val Loss")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Best shuffled_ratio:", study.best_params["shuffled_ratio"])
print("Best validation loss:", study.best_value)
```
